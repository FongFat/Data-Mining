## Equi-width ranges :

```python
min_value  = int(np.floor(df.min().min()))
max_value = int(np.ceil(df.max().max()))
num_bins = 10 # số bins
wrange = max_value - min_value
ivalue = int(np.ceil(wrange/10)) # (max - min)/ số bins
ivalue
```

```python
for i in range(1, num_bins+1):
    intervals = [min_value + (i-1)*ivalue, min_value + i*ivalue]
    print("Bin" + str(i) + " = ", intervals)
```

## Equi-depth ranges:

```python
arr = df.values
arr = arr.astype(dtype=float)
arr.sort()
```

```python
num_bins = 10 # số bins
n = len(arr) 
ivalue = int(np.round(n/num_bins)) # chiều dài data / số bins
for i in range(1, num_bins+1):
    intervals = [] # khởi tạo bin
    for j in range(i*ivalue, (i+1)*ivalue):
        if j >= n:
            break
        intervals.append(arr[j]) # thêm phần tử vào bins
    #print(intervals)
    print("Bin" + str(i) + " : ", len(intervals)) # in ra số giá trị của mỗi bin
```

# 2. PCA

### 1. Chuẩn bị dữ liệu

```python
# Nhập thư viện
import numpy as np
import pandas as pd
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/musk/clean1.data.Z"
# vì file clean1.data.Z không đọc được nên em đọc file clean1.data luôn
df = pd.read_csv("D:\Kì 6\Khai thác dữ liệu\clean1.data", header=None)
```

```python
df.head()
```

```python
# Xóa 2 cột đầu là kiểu chuỗi không thể tính toán
df_v2 = df.drop(df.columns[[0, 1]], axis=1)
df_v2.head()
```

```python
# Tính toán thống kê cơ bản
df_v2.describe()
```

```python
# Chuyển dữ liệu sang ma trận
arr = df_v2.values
arr
```

### 2. Tính mean và lấy từng phần tử của ma trận trừ mean

```python
mean = np.mean(arr, axis=0)
arr = arr - mean
arr
```

### 3. Tìm ma trận hiệp phương sai

```python
cov = arr.T.dot(arr)/arr.shape[0]
cov
```

### 4. Tính trị riêng, vector riêng

```python
from numpy.linalg import eig
eigen_values, eigen_vectors = eig(cov)
```

```python
eigen_values
```

```python
eigen_vectors
```

### 5. Chọn k giá trị riêng lớn nhất để tạo ma trận U

```python
k = 20 # chọn k = 20
select_index = np.argsort(eigen_values)[::-1][:k]
U = eigen_vectors[:, select_index]
```

```python
U
```

### 6. Ánh xạ dữ liệu sang không gian mới

```python
arr_new = arr.dot(U)
arr_new
```
