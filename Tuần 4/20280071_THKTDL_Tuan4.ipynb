# Tự triển khai thuật toán Apriori

```python
# Đọc dữ liệu
df = pd.read_csv("D:\Kì 6\Khai thác dữ liệu\Tuần 4\GroceryStoreDataSet.csv", header=None)
df
```

```python
# Đổi tên cột
df.index.rename('TID', inplace=True)
df.rename(columns={0 : 'item_list'}, inplace=True)
```

```python
trans_df = df.item_list.str.split(',')
trans_df
```

```python
# Các mặt hàng có trong dữ liệu
def itemset(transactions):
    itemsets = set()
    for tran in transactions:
        for item in tran:
            itemsets.add(frozenset([item]))
    return itemsets
```

```python
itemset(trans_df)
```

```python
# Tính support
def calculate_support(itemset, transactions):
    count = 0
    for tran in transactions:
        if itemset.issubset(tran):
            count += 1
    support = count / len(transactions)
    return support, count
```

```python
# Sinh ra các candidate itemset mới từ các frequent itemset kích thước trước
def generate_candidates(frequent_itemsets, k):
    new_candidate = set()
    for itemset1 in frequent_itemsets:
        for itemset2 in frequent_itemsets:
            if itemset1 != itemset2 :
                if len(itemset1.union(itemset2)) == k:
                    new_itemset = itemset1.union(itemset2) # Ghép 2 frequent itemset thành candidate itemset
                    new_candidate.add(new_itemset)

    return new_candidate
```

```python
# Tìm các frequent itemset
def get_frequent_itemsets(transactions, min_support, max_length):

    frequent_itemsets = set()
    support = {}
    count = {}

    # Tìm các frequent_itemsets khởi đầu (1-itemset)
    itemsets = itemset(transactions)
    for item in itemsets:
        supp, cnt = calculate_support(item, transactions)
        
        if supp >= min_support:
            support[item] = supp
            frequent_itemsets.add(item)
            count[item] = cnt
    # Tìm frequent_itemsets có kích thước k (k-itemset)
    k = 2
    while k <= max_length:
        itemsets = generate_candidates(itemset(transactions), k)
        for item in itemsets:
            supp, cnt = calculate_support(item, transactions)
            
            if supp >= min_support:
                support[item] = supp
                frequent_itemsets.add(item)
                count[item] = cnt
        k += 1

    return frequent_itemsets, support, count
```

```python
freq_itemset, support, itemset_count = get_frequent_itemsets(trans_df, 0.2, 2)
p = pd.DataFrame()
p['Frequent_Itemset'] = support.keys()
p['Count'] = itemset_count.values()
p['Support'] = support.values()
p
```

```python
# Tính confidence
def compute_confidence(antecedent, consequent):
    """
    antecedent : mặt hàng trong phần mệnh đề
    consequent : mặt hàng trong phần kết luận
    """
    freq_itemset, support, itemset_count = get_frequent_itemsets(trans_df, 0.2, 2)
    freq = antecedent.union(consequent)
    for itemset in freq_itemset:
        if freq.issubset(itemset) and antecedent.issubset(itemset):
            confidence = support[freq] / support[antecedent]
    return confidence
```

```python
# Tạo các luật kết hợp từ các frequent itemset và lọc các confidence
def generate_strong_rules(frequent_itemset, min_confidence):
    rules = []
    for itemset in frequent_itemset:
        if len(itemset) > 1:
            for item in itemset:
                antecedent = itemset - frozenset([item])
                confidence = compute_confidence(antecedent, frozenset([item]))
                if confidence >= min_confidence:
                    rules.append((antecedent, frozenset([item]), confidence))
    return rules
        
```

```python
min_confidence = 0.5
rules = generate_strong_rules(freq_itemset, min_confidence)
col = ['Left_Hand_Side', 'Right_Hand_Side', 'Confidence']
data = rules
pd.DataFrame(data, columns=col)
```

```python
# Tính Lift
def compute_lift(antecedent, consequent):
    freq_itemset, support, itemset_count = get_frequent_itemsets(trans_df, 0.2, 2)
    freq = antecedent.union(consequent)
    for itemset in freq_itemset:
        if freq.issubset(itemset):
            lift = support[freq] / (support[antecedent] * support[consequent])
    return lift
```

```python
# Tạo các luật kết hợp từ các frequent itemset và lọc các lift
def generate_interesting_rules(frequent_itemset, min_lift):
    rules = []
    for itemset in frequent_itemset:
        if len(itemset) > 1:
            for item in itemset:
                antecedent = itemset - frozenset([item])
                lift = compute_lift(antecedent, frozenset([item]))
                if lift >= min_lift:
                    rules.append((antecedent, frozenset([item]), lift))
                break
    return rules
        
```

```python
min_lift = 1.0
rules = generate_interesting_rules(freq_itemset, min_lift)
col = ['Left_Hand_Side', 'Right_Hand_Side', 'Lift']
data = rules
pd.DataFrame(data, columns=col)
```

# Sử dụng thư viện Apyori

```python
# Xây dựng mô hình Apriori với min_support=0.2
rules = apriori(trans_df, min_support=0.2, max_length=2)
results = list(rules)
results
```

```python
itemset = []
support = []
for item in results:
    itemset.append(item.items)
    support.append(item.support)
pd.DataFrame({'Frequent_Itemsets': itemset, 'Support': support})
```

```python
# Xây dựng mô hình Apriori với min_confidence=0.5
rules = apriori(trans_df, min_support=0.2, min_confidence=0.5, max_length=2)
results = list(rules)
frequent_itemsets = [itemset for itemset in results if len(itemset.items) > 1]
frequent_itemsets
```

```python
left = []
right = []
confidence = []
for item in frequent_itemsets:
    for i in item.ordered_statistics:
        left.append(i.items_base)
        right.append(i.items_add)
        confidence.append(i.confidence)
pd.DataFrame({'Left_Hand_Side': left, 'Right_Hand_Side': right, 'Confidence': confidence})
```

```python
# Xây dựng mô hình Apriori với min_lift=1.1
rules = apriori(trans_df, min_support=0.2, min_lift=1.0, max_length=2)
results = list(rules)
frequent_itemsets = [itemset for itemset in results if len(itemset.items) > 1]
frequent_itemsets
```

```python
itemset = []
lift = []
for item in frequent_itemsets:
    for i in item.ordered_statistics:
        if len(i.items_base) > 0:
            itemset.append(item.items)
            lift.append(i.lift)
            break
print(len(lift))
pd.DataFrame({'Rules': itemset, 'Lift': lift})
```
