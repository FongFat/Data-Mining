## Similarity

```python
def intersection(x, y):
    t = []
    for i in range(df.shape[1]):
      if (i not in (1, 2, 3, 41)):
        if (x[i+41] == y[i+41]):
            t.append(x[i])
    return len(t)

intersection(df_1.loc[0], df_1.loc[1])
```

## IDF

```python
from collections import Counter
MAX_SIZE = 1000000
def IOF(x, y):
    p = 0
    for i in range(df.shape[1]):
        if (i not in (1, 2, 3, 41)):
          cnt = Counter(df_1.loc[:MAX_SIZE, i]) # tần suất xuất hiện
          if (x[i+41] == y[i+41]): # common
            p = p + (cnt[x[i]]/len(df_1.loc[:MAX_SIZE, i]))
          
    return (1/pow(p,2)/df.shape[1])
```

```python
IOF(df_1.loc[0], df_1.loc[1])
```
